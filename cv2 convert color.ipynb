```python
# defining the blues
lower_blue = np.array([110, 50, 50])
upper_blue = np.array([130, 255, 255])
```

```python
# makes a mask to get only blues
mask = cv2.inRange(hsv_bottle, lower_blue, upper_blue)
plt.imshow(mask)
```

```python
# bitwise and takes in src1, src2, mask
# for now we'll use src1 and src2 as the same
results = cv2.bitwise_and(rgb_bottle, rgb_bottle, mask = mask)
plt.imshow(results)
```
